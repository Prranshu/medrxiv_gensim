```python
'''
https://api.medrxiv.org/

The format of the endpoint is https://api.medrxiv.org/details/[server]/[interval]/[cursor]/[format] or https://api.medrxiv.org/details/[server]/[DOI]/na/[format]

where 'interval' can be 1) two YYYY-MM-DD dates separted by '/' and 'cursor' is the start point which defaults to 0 if not supplied, or 2) a numeric value for the N most recent posts, or 3) a numeric with the letter 'd' for the most recent N days of posts.

Where metadata for multiple papers is returned, results are paginated with 100 papers served in a call. The 'cursor' value can be used to iterate through the result.

For instance, https://api.medrxiv.org/details/biorxiv/2018-08-21/2018-08-28/45 will output 100 results (if that many remain) within the date range of 2018-08-21 to 2018-08-28 beginning from result 45 for biorxiv. https://api.medrxiv.org/details/medrxiv/2020-03-21/2020-03-24/45 will output 100 results (if that many remain) within the date range of 2020-03-21 to 2020-03-24 beginning from result 45 for medrxiv.

https://api.medrxiv.org/details/[server]/[DOI]/na/[format] returns detail for a single manuscript. For instance, https://api.medrxiv.org/details/medrxiv/10.1101/2020.09.09.20191205 will output metadata for the medrxiv paper with DOI 10.1101/339747.

The 'messages' array in the output provides information about what is being displayed, including cursor value, count of all items and count of new papers for the requested interval.

Available formats are JSON (json) and XML (OAI-PMH XML).

The following metadata elements are returned:

doi
title
authors
author_corresponding
author_corresponding_institution
date
version
type
license
category
jats xml path
abstract
published
server
'''
```

```python
url_list = []
for i in range(250000, 300000, 100):
    url = 'https://api.medrxiv.org/details/biorxiv/2013-08-01/2023-08-01/{}'.format(i)
    url_list.append(url)
l = list(range(0, 500, 10))
a = []
for i in url_list:
    if url_list.index(i) in l:
        time.sleep(5)
        print('Slept for 5s.')
    b = requests.get(i)
    c = b.json()
    d = c['collection']
    e = pd.DataFrame(d)
    a.append(e)
    print(f'{i} appended')
df = pd.concat(a)
medrxiv_6 = df.to_csv('./medrxiv_6.csv')
pd.read_csv('medrxiv_6.csv')
```
